```python
import pandas as pd 
```

```python
testData = {
    'name' : ['Somesh','Samit','Ninad', 'Prasad', 'Om'],
    'height' : [6.3, 6.2, 5.2, 5.4, 5.8]
}
```

```python
df = pd.DataFrame(testData)
```

```python
df
```

```python
# these outliers are on the higher end
max_threshold = df['height'].quantile(0.95)
max_threshold
```

```python
 df[df['height'] > max_threshold]
```

```python
# these outliers are on the lower end 
min_threshold = df['height'].quantile(0.05)
min_threshold 
```

```python
df[df['height'] < min_threshold]
```

```python
# we can do the above mentioned in one go itself 
df2 = df[(df['height'] < max_threshold) & (df['height'] > min_threshold)]
df2 # now df2 has almost no outliers remaining!
```
